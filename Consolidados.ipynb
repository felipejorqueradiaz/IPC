```python
from google.colab import drive  #Para enlazar a Drive
import pandas as pd             #Para manejar la base de datos
import glob, os                 #Para leer archivos de drive
import datetime                 #Para manejar fechas
import numpy as np
import time
import re

drive.mount('/content/drive')
```

```python
path = 'drive/MyDrive/IPC Bienes/'
productos = ['Celulares',
             'TV',
             'Impresora',
             'Lavadora - Secadora',
             'Computador', 
             'Refrigerador'
             ]
```

```python
fecha_inicio = '2021-05-01'
fecha_final = '2021-05-31'
mes = 'Mayo'
agno = '2021'
```

```python
for producto in productos:
  db = {}
  i=1
  for filename in glob.glob(os.path.join(path+producto, '*.csv')):
    archivo = re.findall(pattern='--(.*).csv', string=filename)[0]
    if (archivo >= fecha_inicio) & (archivo <= fecha_final):
      df = pd.read_csv(filename, encoding='utf-8', parse_dates=True)

      try:
        df = df.rename({'Precio_Normal': 'Precio Normal'}, axis=1)
      except:
        pass


      try: df['Precio Normal'] = df['Precio Normal'].replace('[\$,]','',regex=True).replace('[\.,]','',regex=True).replace('[\",]','',regex=True).astype(int)
      except:
        for j, row in df.iterrows():
          valor = row['Precio Normal']
          try:
            df.at[j, 'Precio Normal'] = valor.replace('[\$,]','',regex=True).replace('[\.,]','',regex=True).replace('[\",]','',regex=True).astype(int)
          except:
            df.at[j, 'Precio Normal'] = np.nan
          
      df = df[df['Precio Normal'].notna()]
      df['Precio Normal'] = df['Precio Normal'].astype(int)
      db[i] = df
      i=i+1
  x = pd.concat(db.values())
  x.to_csv('drive/MyDrive/IPC Bienes/Consolidados/'+producto+'/Scrapers/'+mes+'-'+agno+'.csv', index=False)
```
